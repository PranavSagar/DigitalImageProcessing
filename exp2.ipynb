Average weighted method of converting rgb to grey

```python
#to seperate bgr components using python function
b = img[:,:,0]
g = img[:,:,1]
r = img[:,:,2]
grey = r * 0.11+ g * 0.59 + b * 0.39
cv2_imshow(grey)
```

```python
#in python we have function to convert a rgb to gray model
gray2 = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray2)
```

```python
gray3 = cv2.cvtColor(gray2,cv2.COLOR_GRAY2RGB)
cv2_imshow(gray3)
```

```python
print(gray2.shape)
print(gray3.shape)
```

```python
#to seperate bgr components using python function
b,g,r = cv2.split(img)
gray4 = r * 0.11+ g * 0.59 + b * 0.39
cv2_imshow(gray4)
```
