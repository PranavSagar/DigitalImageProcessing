```python
import numpy as np
```

```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread("dipexp3.jpg")
```

```python
cv2_imshow(img)
```

```python
c = 255/(1+np.max(img))
```

```python
for i in range(1,5):
  newimg = c*np.log(1+img)*i*10
  cv2_imshow(newimg)
```

```python
gama = 0.5
while(gama!=1.5):
  newimg = c*img**gama
  gama = gama + 0.25
  cv2_imshow(newimg)
```
