```python
#import lib
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread("/content/drive/MyDrive/Sem6th/DIP/Exp3/exp3c.jpg")
```

```python
cv2_imshow(img)
```

```python
#cv2 read image in BGR color model
#covert BGR to GRAY model
grayimg = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2_imshow(grayimg)
```

```python
m=np.max(grayimg)
print(m)
```

```python
d = {}
sum = 0
for i in range(0,m+1):
  a=np.sum(grayimg==i)
  sum = sum + a
  d[i] = a
  
```

```python
print(d)
```

```python
pdf = {}
size = grayimg.size
for i in range(0,m+1):
  pdf[i] = d[i]/size
  # cdf[i] = cdf[i] + cdf[i-1]
```

```python
print(pdf)
```

```python
# making cdf 
cdf = {}
cdf[0] = pdf[0]
for i in range(1,m+1):
  cdf[i] = cdf[i-1]+ pdf[i]
```

```python
print(cdf)
```

```python
roundcdf = {}
for i in range(0,m+1):
   roundcdf[i] = round(cdf[i]*(m+1))
```

```python
print(roundcdf)
```

```python
cv2_imshow(grayimg)
```

```python
row = img.shape[0]
col = img.shape[1]
for k in range(0,m+1):
  for i in range(row):
    for j in range(col):
      if grayimg[i,j] == k:
        grayimg[i,j] = roundcdf[k]
```

```python
cv2_imshow(grayimg)
```
